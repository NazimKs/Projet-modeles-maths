```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Données
results_approche_1 = [[0.265,0.042,0.008,0.098,0.056],[0.059,0.366,0.581,0.114,0.245],[4.7,10.2,25.8,3.44,1.02],[300,44,95.6,140,80.1],[300,300,300,300,300]]
results_approche_2 = [[0.101,0.448,0.133,0.166,0.286],[4.71,13.4,10,43,2.12],[227,79.6,300,106,14.7],[300,300,300,300,300],[69.3,84.5,97.3,52.7,51.3]]

# Convertion en numpy arrays pour la manipulation
results_approche_1 = np.array(results_approche_1)
results_approche_2 = np.array(results_approche_2)

# Tailles des instances
sizes = [10, 20, 30, 50, 100]

# Moyennes des temps de résolution pour chaque taille
mean_approche_1 = results_approche_1.mean(axis=1)
mean_approche_2 = results_approche_2.mean(axis=1)

# Graphique en barres
plt.figure(figsize=(12, 6))
x = np.arange(len(sizes))
width = 0.35

plt.bar(x - width/2, mean_approche_1, width, label='Génération de Contraintes')
plt.bar(x + width/2, mean_approche_2, width, label='Formulation MTZ')

plt.xlabel('Nombre de Villes')
plt.ylabel('Temps Moyen de Résolution (s)')
plt.title('Comparaison des Temps de Résolution Moyens')
plt.xticks(x, sizes)
plt.legend()

plt.savefig('evolution_temps_resolution.png')

plt.show()


```
